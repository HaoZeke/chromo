```python
from impy.models import Pythia8
from impy.kinematics import CenterOfMass
from impy.constants import TeV, GeV

ekin = CenterOfMass(10 * GeV, "proton", "proton")

gen = Pythia8(ekin, 2)
```

```python
events = list(gen(2))
```

```python
event = events[1]
event
```

```python
import numpy as np

quarks_and_diquarks_and_gluons = {
 1,
 2,
 3,
 4,
 5,
 6,
 21,
 1103,
 2101,
 2103,
 2203,
 3101,
 3103,
 3201,
 3203,
 3303,
 4101,
 4103,
 4201,
 4203,
 4301,
 4303,
 4403,
 5101,
 5103,
 5201,
 5203,
 5301,
 5303,
 5401,
 5403,
 5503
}

ma = True
for x in quarks_and_diquarks_and_gluons:
    ma &= np.abs(event.pid) != x

event2 = event.copy()
n = len(event2)

for i, (a, b) in enumerate(event2.parents):
    if a == 0:
        continue
    while not ma[a - 1]:
        a = event2.parents[a - 1][0]
    if a < 3:
        event2.parents[i] = (1, 2)
    else:
        event2.parents[i] = (a, 0)

# for i in range(len(event2)):
#     print(event2.parents[i], event2.pid[i])

indices = np.arange(len(event2)) + 1

indices1 = indices[ma]
mapping = {old: i + 1 for i, old in enumerate(indices1)}

par = event2.parents[ma]
for i, (a, b) in enumerate(par):
    if a > 0:
        par[i, 0] = mapping[a]

event2 = event2[ma]
event2.parents = par

for i in range(len(event2)):
    print(event2.parents[i], event2.pid[i], i, indices1[i])

event2
```
